```python
from pandas import DataFrame

def process_data(data):
    items = []
    print(data)
    for d in data['items']:
        item = {'publishedAt': d['snippet']['publishedAt'], 
            'channelId': d['snippet']['channelId'], 
            'Title': d['snippet']['title'], 
            'description': d['snippet']['description']
            }
        items.append(item)

    df = DataFrame(items) 
    return df
```

```python
import pandas as pd
import gspread
def get_worksheet(sheet:gspread.spreadsheet, 
                  ws_name: str) -> gspread.worksheet:
    try: 
        worksheet = sheet.worksheet(ws_name)
    except Exception as e: 
        worksheet = sheet.add_worksheet(ws_name, rows=100, cols=20)
    return worksheet
def upload_spreadsheet(df: pd.DataFrame, 
                       ws: gspread.worksheet) -> bool:
    try:
        print(df)
        ws.update([df.columns.values.tolist()] + df.values.tolist())
        return True
    except Exception as err:
        print(err)
        return False
```
